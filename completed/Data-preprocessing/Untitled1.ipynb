```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Data.csv')
```

```python
df
```

```python
from sklearn.preprocessing import Imputer
```

```python
imputer = Imputer(missing_values = 'NaN', strategy = 'mean', axis = 0)
```

```python
imputer = imputer.fit(df[['Age', 'Salary']])
```

```python
df[['Age', 'Salary']] = imputer.transform(df[['Age', 'Salary']])
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
```

```python
labelencoder_X = LabelEncoder()
```

```python
df['Country'] = labelencoder_X.fit_transform(df['Country'])
```

```python
df.head()
```

```python
onehotencoder = OneHotEncoder(categorical_features = [0])
```

```python
X = pd.DataFrame(onehotencoder.fit_transform(df[['Country','Age', 'Salary']]).toarray())
```

```python
X.describe()
```

```python
sc_X = StandardScaler()
df1 = pd.DataFrame(sc_X.fit_transform(X))
```
